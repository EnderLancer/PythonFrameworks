### Завдання 1 ###

```python
import pandas as pd

index1 = pd.MultiIndex.from_tuples([('Egypt', 2018), ('Egypt', 2018), ('Egypt', 2019),
                                    ('Turkey', 2018), ('Turkey', 2019), ('Turkey', 2020)], names=['Country', 'Year'])
ser1 = pd.Series(['Ivan Ivanov', 'Petro Petrov', 'Katerina Kulbaba', 
                  'Pavlo Sydorenko', 'Ivan Dubenko', 'Makar Khmara'], index=index1)
df1 = pd.DataFrame({'Name': ser1,
                    'Agency': ['Sun', 'Moon', 'Star', 
                               'Sun', 'Star', 'Star']})
df1
```

### Завдання 2 ###

```python
df2 = pd.DataFrame({'Country': ['Egypt', 'Egypt', 'Turkey', 'Turkey', 'Ukraine', 'Ukraine'],
                    'City': ['Cairo', 'Luxor', 'Istanbul', 'Izmir', 'Lviv', 'Kharkiv'],
                    'Days to stay': [5, 4, 5, 3, 4, 3],
                    'Transport': ['Train', 'Bus', 'Airplane', 'Airplane', 'Boat', 'Car'],
                    'Tour cost': [123423, 234234, 456454, 734533, 734534, 644522]})
df2
```

### Завдання 3 ###

```python
df = pd.merge(df1.reset_index(), df2, on='Country', how='outer').set_index('Country')
df
```

### Завдання 4 ###

```python
df = df.fillna(method='ffill')
df
```

### Завдання 5 ###

```python
df.groupby('Country').agg({'Tour cost': 'mean'}).sort_values("Tour cost", ascending=True).head(1)
```

### Завдання 6 ###

```python
df['Cost per day'] = df['Tour cost'] / df['Days to stay']
df
```

### Завдання 7 ###
Враховуючи завдання, то скоріш малось на увазі вивести Name агенства с туром, бо звичайне ім'я покупця було б не доцільно виводити.

```python
df.groupby(['Country', 'Agency']).agg({"Cost per day": 'mean'}).sort_values("Cost per day", ascending=False).head(2).tail(1)
```
