```python
import pandas as pd
import re

books = pd.read_csv('books.csv')
books
```

```python
def count_vowel(string):
    string = "".join(re.findall('[.*[aeiou]+]*', str(string).lower())) 
    return len(string) 
copy2 = books.copy()
copy2['VowelsInName'] = copy2.apply(lambda x: count_vowel(x.Author), axis=1)
copy2
```

```python
books.loc[books['Author'].str.extract('.*([aeiou]{3})').dropna().index.values, 'Title':'Publisher']
```

```python
def is_digit(string):
    string = "".join(re.findall('[.*[\d]+]*', str(string).lower())) 
    return (1 if len(string) > 0 else 0)
copy4 = books.copy()
copy4['HaveDigitInName'] = copy4.apply(lambda x: is_digit(x.Title), axis=1)
copy4.head()
```

```python
copy5 = books.copy()
copy5['TitleLength'] = copy5['Title'].str.len()
copy5[['Title', 'TitleLength']].sort_values(by=['TitleLength'], ascending=False).head(1)
```

```python
copy6 = books.copy().dropna()
copy6['IsAuthorWithA'] = copy6['Author'].str.startswith('A')
copy6[copy6.IsAuthorWithA].head()
```

```python
def find_double_title(string):
    return (len(string.split()) == 2) 
copy7 = books.copy()
copy7['IsBookWithDoubleTitle'] = copy7.apply(lambda x: find_double_title(x.Title), axis=1)
copy7[copy7.IsBookWithDoubleTitle].head()
```

```python
def is_three_y(string):
    string = "".join(re.findall('[?*y]*', str(string).lower()))
    return (len(string) == 3) 
copy8 = books.copy()
copy8['IsThreeYInName'] = copy8.apply(lambda x: is_three_y(x.Author), axis=1)
copy8[copy8.IsThreeYInName]
```
